```python
# READING DATA FROM MULTIPLE DIRECTORY
import glob
```

```python
for directory in glob.glob('datasets/multi-directory/*'):
    print(directory)
```

```python
# iterate through each file
for dir in glob.glob('datasets/multi-directory/*'):
    # iterate each file
    for files in glob.glob(dir + '/*'):
        print(files)
```

```python
# Concatenate the files
```

```python
# list to store the dataframe
dataFrameList = []

# iterate each folder
for dir in glob.glob('datasets/multi-directory/*'):
    # iterate files
    for files in glob.glob(dir + '/*'):
        # combine the files by array
        dataFrameList.append(pd.read_csv(files))
```

```python
dataFrameList
```

```python
len(dataFrameList)
```

```python
# concate the data
finalData = pd.concat(dataFrameList)
```

```python
finalData.head()
```

```python
finalData.shape
```

```python
# STORE THE CSV
dataMergedPrev = pd.read_csv('datasets/merged_big_mart_sales.csv')
dataMergedPrev.shape
```

```python
# Store the finalData to CSV file
finalData.to_csv('datasets/merged_big_mart_sales.csv', index=False)
```
