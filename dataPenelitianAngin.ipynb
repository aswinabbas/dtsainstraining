```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import weibull_min
from scipy.optimize import minimize

# ======================================
# 1. Generate Sample Data (Hourly Jan-Dec)
# ======================================
np.random.seed(42)
hours_per_year = 24 * 365
data = weibull_min.rvs(2, scale=7, size=hours_per_year)  # k=2, c=7

# ======================================
# 2. Weibull MLE Fitting
# ======================================
def neg_log_likelihood(params, data):
    k, c = params
    if k <= 0 or c <= 0:
        return np.inf  # Return infinity for invalid parameters
    return -np.sum(np.log(k) - k*np.log(c) + (k-1)*np.log(data) - (data/c)**k)

# Initial guess for k and c
initial_params = [1.5, 6.0]

# Minimize negative log-likelihood
result = minimize(neg_log_likelihood, 
                 initial_params, 
                 args=(data,),
                 bounds=[(0.1, 10), (0.1, 20)])  # k: 0.1-10, c: 0.1-20

k_mle, c_mle = result.x
print(f"Estimated parameters: k = {k_mle:.2f}, c = {c_mle:.2f}")

# ======================================
# 3. Visualization
# ======================================
plt.figure(figsize=(12, 6))

# Histogram data
plt.hist(data, bins=50, density=True, alpha=0.6, label='Hourly Wind Speed Data')

# Weibull PDF with fitted parameters
x = np.linspace(0, 20, 200)
plt.plot(x, weibull_min.pdf(x, k_mle, scale=c_mle), 
         'r-', lw=2, label=f'Weibull Fit (k={k_mle:.2f}, c={c_mle:.2f})')

plt.title('Hourly Wind Speed Distribution (Jan-Dec) with Weibull Fit')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# ======================================
# 4. Additional Analysis
# ======================================
# Calculate mean wind speed from Weibull parameters
from scipy.special import gamma
mean_speed = c_mle * gamma(1 + 1/k_mle)
print(f"Mean wind speed: {mean_speed:.2f} m/s")

# Calculate wind power density
rho = 1.225  # Air density (kg/m³)
power_density = 0.5 * rho * c_mle**3 * gamma(1 + 3/k_mle)
print(f"Wind power density: {power_density:.2f} W/m²")
```
