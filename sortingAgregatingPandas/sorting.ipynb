```python
import pandas as pd
```

```python
# Creta sample dataframe
dataFrame = pd.DataFrame({
    'roll_no': [102, 101, 104, 103, 105],
    'name': ['Aravind', 'Rahul', 'Prateek', 'Piyuesh', 'Kartik'],
    'grade': ['B', 'B', 'A', 'C', 'A'],
    'marks': [15, 15, 20, 4, 22],
    'city': ['Gurugram', 'Delhi', 'Delhi', 'Gurugram', 'Hyderabad']
})

dataFrame
```

```python
# Sort dataframe by by grade of the student
dataFrame.sort_values(by="grade")
```

```python
dataFrame.sort_values(by=["roll_no"])
```

```python
# sort dataframe by more than one parameters
dataFrame.sort_values(by=["grade", "marks"], ascending=[True, False])
```

```python
dataFrame
```

```python
# SAVE the sorted value by using inplace
dataFrame.sort_values(by=["grade", "marks"], ascending=[True, False], inplace=True)
```

```python
dataFrame
```

```python
# reset index
dataFrame.reset_index()
```

```python
# replace the index coulomn
dataFrame.reset_index(inplace=True, drop=True)
dataFrame
```
