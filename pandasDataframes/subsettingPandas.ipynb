----------------------------------------SUBSETTING: VALUE BASED-----------------------------------------------------

```python
data = pd.read_csv('./datasets/big_mart_sales.csv')
data.head()
```

```python
# filter rows with condition
data[data.Outlet_Establishment_Year == 1987]

# it would be same with 
# data.loc[data.Outlet_Establishment_Year == 1987]
```

```python
# check by multiple condition # wiht & sign (only 2009 AND Medium would be showed)
data[(data.Outlet_Establishment_Year == 2009) & (data.Outlet_Size == 'Medium')] 
```

```python
# check by multiple condition # wiht | sign (only 2009 OR Medium would be showed)
data[(data.Outlet_Establishment_Year == 2009) | (data.Outlet_Size == 'Medium')] 
```

```python
# check by multiple condition # wiht != sign (data except 2009)
data[(data.Outlet_Establishment_Year != 2009)] 
```

```python
# FILTER FOR LIST OF VALUES
```

```python
# Get rows for 1987, 1988, 1999
data[(data.Outlet_Establishment_Year == 1987) | (data.Outlet_Establishment_Year == 1988) | (data.Outlet_Establishment_Year == 1999)]
```

```python
# Get rows for 1987, 1988, 1999 by using isin
data[data.Outlet_Establishment_Year.isin([1987, 1988, 1999])]
```

```python
# Select data for specific columns
```

```python
columns = ['Item_Identifier', 'Item_MRP', 'Outlet_Establishment_Year', 'Outlet_Size']
data[columns]
```

```python
# select spesific value and specific coulum
data[(data.Outlet_Establishment_Year == 2009) | (data.Outlet_Size == 'High')][columns]
```

```python
# using loc
data.loc[(data.Outlet_Establishment_Year == 2009) | (data.Outlet_Size == 'High'), columns]
```
